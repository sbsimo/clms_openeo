```python
import json
import glob
import os

res = {}
for json_path in glob.glob('mystorage/results/results17/lai_biopar_20*/timeseries.json'):
    with open(json_path, 'r') as f:
        print(os.path.dirname(json_path))
        json_res = json.load(f)
        res[os.path.dirname(int(json_path)[-4:])] = json_res[0][0]

print(res)
```

```python
import matplotlib.pyplot as plt

plt.plot(res.keys(), res.values())
plt.show()
```
